–ù–∞ —Å–∞–º–æ–º –¥–µ–ª–µ —á–µ–º –±–æ–ª—å—à–µ –¥–∞–Ω–Ω—ã—Ö –æ—Å—Ç–∞–µ—Ç—Å—è, —Ç–µ–º –ª—É—á—à–µ, –ø–æ—ç—Ç–æ–º—É –ª—É—á—à–µ —É–±–∏—Ä–∞—Ç—å –∏–º–µ–Ω–Ω–æ –≤—ã–±–∏–≤–∞—é—â–∏–µ—Å—è –∑–Ω–∞—á–µ–Ω–∏—è, –µ—Å–ª–∏ –∏—Ö –º–Ω–æ–≥–æ, —Ç–æ –∑–Ω–∞—á–∏—Ç —É –Ω–∞—Å –¥–∞–Ω–Ω—ã–µ —Ä–∞–∑–º–∞–∑–∞–Ω—ã, –∞ —ç—Ç–∏ –∑–Ω–∞—á–µ–Ω–∏—è –Ω–æ—Ä–º–∞–ª—å–Ω—ã–µ(–∞ –Ω–µ –≤—ã–±–∏–≤–∞—é—â–∏–µ—Å—è).

–Ø –≤ —Ä–∞–±–æ—Ç–µ –ø–æ–ª—å–∑—É—é—Å—å –º–µ—Ç–æ–¥–æ–º [quantile](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.quantile.html) –∏ —É–±–∏—Ä–∞—é —Å –æ–±–µ–∏—Ö —Å—Ç–æ—Ä–æ–Ω –¥–∞–Ω–Ω—ã–µ, –∫–æ—Ç–æ—Ä—ã–µ –≤—ã—Ö–æ–¥—è—Ç –∑–∞ 1-3%(—Ç–æ –µ—Å—Ç—å –º–µ–Ω–µ–µ 0.03 –∏ –±–æ–ª–µ–µ 0.97 –ø–µ—Ä—Ü–µ–Ω—Ç–∏–ª—è), –µ—Å–ª–∏ —Å–¥–µ–ª–∞—Ç—å —Ç–∞–∫–æ–µ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –ø–æ –Ω–µ—Å–∫–æ–ª—å–∫–∏–º –∫–æ–ª–æ–Ω–∫–∞–º, —Ç–æ –∫–∞–∫ —Ä–∞–∑ –∏ –ø–æ–ª—É—á–∏—Ç—Å—è –æ–∫–æ–ª–æ 10%. –ù–∏–∂–µ –ø–æ–∫–∞–∑–∞–ª.

````python
# –ö–æ–¥ —Ä–µ–≤—å—é–µ—Ä–∞

low_percentele = 0.01
hihg_percentele = 0.99

last_price_low = data['last_price'].quantile(low_percentele)
last_price_high = data['last_price'].quantile(hihg_percentele)

print(f'–ù–∏–∂–Ω—è—è –≥—Ä–∞–Ω–∏—Ü–∞ last_price {last_price_low}')
print(f'–ù–∏–∂–Ω—è—è –≥—Ä–∞–Ω–∏—Ü–∞ last_price {last_price_high}')

print(
    '–ü—Ä–æ—Ü–µ–Ω—Ç –æ—Ç–±—Ä–æ—à–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π = ',
    round(
        100 - len(data.query('@last_price_low <= last_price <= @last_price_high'))
        / len(data['last_price']) * 100,
        2
    ),
    '%'
)
````
